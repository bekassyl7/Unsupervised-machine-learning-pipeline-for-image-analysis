```python
import os
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
import matplotlib as mpl
from pylab import cm

mpl.rcParams.update(mpl.rcParamsDefault)

OUTPUT_DATASET = "Results_Classification_Cross_Validation"
RESULTS = "Results_Classification_Cross_Validation"
NImages = ["NX_NY_Images_1", "length_125","length_100","NX_NY_Images_2","NX_NY_Images_3","NX_NY_Images_4",
           "NX_NY_Images_5","NX_NY_Images_6","NX_NY_Images_7",
            "NX_NY_Images_8", "NX_NY_Images_9","NX_NY_Images_10"]

Length_Scale = ['150','125','100','75','50','37.5','30','25','21.4','18.75','16.67','15']

file = os.path.sep.join([RESULTS, "Accuracy.txt"])
L=0

results_DIR = os.path.sep.join([RESULTS])
f = open(file,"r")
LenScale = []
CV_fold = []
perp = []
classification_accuracy = []
clustering_accuracy = []
for val in f:
    line = val.split(',')
    LenScale.append(line[0])
    CV_fold.append(line[1])
    perp.append(line[2])
    classification_accuracy.append(line[3])
    clustering_accuracy.append(line[4])

LenScale = LenScale[1:]
CV_fold = CV_fold[1:]
perp = perp[1:]
classification_accuracy = classification_accuracy[1:]
clustering_accuracy = clustering_accuracy[1:]

data = np.empty([len(LenScale),5])
data[:,0] = LenScale
data[:,1] = CV_fold
data[:,2] = perp       
data[:,3] = classification_accuracy
data[:,4] = clustering_accuracy

#%%
count=0
final_data_all_CV=np.empty([65,5])
j=0
for i in data:
    if count % 9 == 0:
        ind_max_clus_acc = np.argwhere(data[count:count+9,4] == np.amax(data[count:count+9,4]))
        ind_max_class_acc = np.argmax(data[ind_max_clus_acc,3])
        final_data_all_CV[j] = data[ind_max_clus_acc[ind_max_class_acc]+count,:]
        j += 1
    count += 1 
  
count=0
final_data_CV1=np.empty([13,5])
final_data_CV2=np.empty([13,5])
final_data_CV3=np.empty([13,5])
final_data_CV4=np.empty([13,5])
final_data_CV5=np.empty([13,5])

final_data_CV1 = final_data_all_CV[range(0,65,5)]
final_data_CV2 = final_data_all_CV[range(1,65,5)]
final_data_CV3 = final_data_all_CV[range(2,65,5)]
final_data_CV4 = final_data_all_CV[range(3,65,5)]
final_data_CV5 = final_data_all_CV[range(4,65,5)]



# Calculate the average
final_data_CV_clustering_acc_mean = np.mean([final_data_CV1[:,4],final_data_CV2[:,4],
                                        final_data_CV3[:,4],final_data_CV4[:,4],
                                        final_data_CV5[:,4]], axis=0)

# # Calculate the standard deviation
final_data_CV_clustering_acc_std = np.std([final_data_CV1[:,4],final_data_CV2[:,4],
                                        final_data_CV3[:,4],final_data_CV4[:,4],
                                        final_data_CV5[:,4]], axis=0)

# Calculate the average
final_data_CV_class_acc_mean = np.mean([final_data_CV1[:,3],final_data_CV2[:,3],
                                        final_data_CV3[:,3],final_data_CV4[:,3],
                                        final_data_CV5[:,3]], axis=0)

# # Calculate the standard deviation
final_data_CV_class_acc_std = np.std([final_data_CV1[:,3],final_data_CV2[:,3],
                                        final_data_CV3[:,3],final_data_CV4[:,3],
                                        final_data_CV5[:,3]], axis=0)


#%%
# Edit the font, font size, and axes width
mpl.rcParams['font.family'] = 'Avenir'
plt.rcParams['font.size'] = 24
plt.rcParams['axes.linewidth'] = 3
# Generate 2 colors from the 'tab10' colormap
colors = cm.get_cmap('tab10', 2)

# Create figure and add axes object
fig = plt.figure(figsize=(8, 5))
ax = fig.add_axes([0, 0, 1, 1])
# Edit the major and minor ticks of the x and y axes
ax.xaxis.set_tick_params(which='major', size=10, width=2, direction='in', top=False)
ax.xaxis.set_tick_params(which='minor', size=10, width=2, direction='in', top=False)
ax.yaxis.set_tick_params(which='major', size=10, width=2, direction='in', right=False)
ax.yaxis.set_tick_params(which='minor', size=7, width=2, direction='in', right=False)

# Plot and show our data
ax.errorbar(final_data_CV1[:,0],final_data_CV_clustering_acc_mean[:],
        yerr=final_data_CV_clustering_acc_std, marker='o',mfc=colors(0),
         ecolor='black', ms=10,elinewidth=2,capsize=10,barsabove=True,
        linewidth=3, color=colors(0), label='Train')
ax.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(10))
ax.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.05))
# Add the x and y-axis labels
ax.set_xlabel(r'Length $(\mathregular{\mu}m)$', labelpad=10)
ax.set_ylabel('Accuracy', labelpad=10)
ax.tick_params(axis='x', which='minor', bottom=True)

# Add legend to plot
ax.legend(bbox_to_anchor=(1, 0), loc="lower right", frameon=False, fontsize=20)
plt.ylim(0.5,1.02)
plt.tight_layout()
out_name = "Accuracy_Train" + '.png'
plt.savefig(os.path.sep.join([results_DIR, out_name]),bbox_inches="tight")
plt.close()

#%%
# Create figure and add axes object
fig1 = plt.figure(figsize=(8, 5))
ax1 = fig1.add_axes([0, 0, 1, 1])
# Edit the major and minor ticks of the x and y axes
ax1.xaxis.set_tick_params(which='major', size=10, width=2, direction='in', top=False)
ax1.xaxis.set_tick_params(which='minor', size=10, width=2, direction='in', top=False)
ax1.yaxis.set_tick_params(which='major', size=10, width=2, direction='in', right=False)
ax1.yaxis.set_tick_params(which='minor', size=7, width=2, direction='in', right=False)

# Plot and show our data
ax1.errorbar(final_data_CV1[:,0],final_data_CV_class_acc_mean[:],
        yerr=final_data_CV_clustering_acc_std, marker='o',mfc=colors(0),
         ecolor='black', ms=10,elinewidth=2,capsize=10,barsabove=True,
        linewidth=3, color=colors(0), label='Test')

ax1.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(10))
ax1.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.05))
# Add the x and y-axis labels
ax1.set_xlabel(r'Length $(\mathregular{\mu}m)$', labelpad=10)
ax1.set_ylabel('Accuracy', labelpad=10)
ax1.tick_params(axis='x', which='minor', bottom=True)

# Add legend to plot
ax1.legend(bbox_to_anchor=(1, 0), loc="lower right", frameon=False, fontsize=20)
plt.ylim(0.5,1.02)
plt.tight_layout()
out_name = "Accuracy_Test" + '.png'
plt.savefig(os.path.sep.join([results_DIR, out_name]),bbox_inches="tight")
plt.close()
```
